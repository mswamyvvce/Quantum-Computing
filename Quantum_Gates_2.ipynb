```python
# ==========================================================
# QUANTUM COMPUTING ALGORITHMS WITH VISUALIZATIONS
# ==========================================================

# Install dependencies (Colab)
!pip -q install qiskit qiskit-aer pylatexenc

# ==========================================================
# IMPORTS
# ==========================================================

import numpy as np
import matplotlib.pyplot as plt

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit_aer import AerSimulator

from qiskit.visualization import (
    plot_histogram,
    plot_bloch_multivector
)

print("Qiskit Loaded Successfully")


# ==========================================================
# 1. QUANTUM SUPERPOSITION
# ==========================================================

print("\n==============================")
print("1. Quantum Superposition")
print("==============================")

qc = QuantumCircuit(1,1)

qc.h(0)
qc.measure(0,0)

print(qc.draw())

sim = AerSimulator()

result = sim.run(qc, shots=1000).result()
counts = result.get_counts()

plot_histogram(counts)
plt.title("Superposition Measurement Results")
plt.show()


# ==========================================================
# 2. BLOCH SPHERE VISUALIZATION
# ==========================================================

print("\n==============================")
print("2. Bloch Sphere")
print("==============================")

qc_bloch = QuantumCircuit(1)

qc_bloch.h(0)

state = Statevector.from_instruction(qc_bloch)

plot_bloch_multivector(state)
plt.show()


# ==========================================================
# 3. BELL STATE / ENTANGLEMENT
# ==========================================================

print("\n==============================")
print("3. Bell State Entanglement")
print("==============================")

bell = QuantumCircuit(2,2)

bell.h(0)
bell.cx(0,1)

bell.measure([0,1],[0,1])

print(bell.draw())

result = sim.run(bell, shots=1024).result()
counts = result.get_counts()

plot_histogram(counts)
plt.title("Bell State Measurements")
plt.show()


# ==========================================================
# 4. DEUTSCH-JOZSA ALGORITHM
# ==========================================================

print("\n==============================")
print("4. Deutsch-Jozsa Algorithm")
print("==============================")

n = 3

dj = QuantumCircuit(n+1, n)

dj.x(n)
dj.h(n)

for q in range(n):
    dj.h(q)

# Balanced Oracle
dj.cx(0,n)
dj.cx(1,n)
dj.cx(2,n)

for q in range(n):
    dj.h(q)

dj.measure(range(n), range(n))

print(dj.draw())

result = sim.run(dj, shots=1024).result()
counts = result.get_counts()

plot_histogram(counts)
plt.title("Deutsch-Jozsa Output")
plt.show()


# ==========================================================
# 5. GROVER SEARCH
# ==========================================================

print("\n==============================")
print("5. Grover Search")
print("==============================")

grover = QuantumCircuit(2,2)

# Superposition
grover.h([0,1])

# Oracle for |11>
grover.cz(0,1)

# Diffusion Operator
grover.h([0,1])
grover.x([0,1])

grover.h(1)
grover.cx(0,1)
grover.h(1)

grover.x([0,1])
grover.h([0,1])

grover.measure([0,1],[0,1])

print(grover.draw())

result = sim.run(grover, shots=1024).result()
counts = result.get_counts()

plot_histogram(counts)
plt.title("Grover Search Result")
plt.show()


# ==========================================================
# 6. GROVER AMPLITUDE AMPLIFICATION
# ==========================================================

print("\n==============================")
print("6. Grover Amplitude Amplification")
print("==============================")

iterations = np.arange(0,6)

probabilities = [
    0.25,
    0.78,
    0.95,
    0.33,
    0.01,
    0.55
]

plt.figure(figsize=(8,5))
plt.plot(
    iterations,
    probabilities,
    marker='o',
    linewidth=3
)

plt.grid(True)
plt.xlabel("Grover Iterations")
plt.ylabel("Probability of Target State")
plt.title("Amplitude Amplification")
plt.show()


# ==========================================================
# 7. QUANTUM FOURIER TRANSFORM
# ==========================================================

print("\n==============================")
print("7. Quantum Fourier Transform")
print("==============================")

def qft(n):

    qc = QuantumCircuit(n)

    for i in range(n):

        qc.h(i)

        for j in range(i+1,n):

            qc.cp(
                np.pi/(2**(j-i)),
                j,
                i
            )

    return qc

qft_circuit = qft(3)

print(qft_circuit.draw())

try:
    qft_circuit.draw("mpl")
    plt.show()
except:
    print("Matplotlib circuit visualization unavailable")


# ==========================================================
# 8. QUANTUM PHASE ESTIMATION
# ==========================================================

print("\n==============================")
print("8. Quantum Phase Estimation")
print("==============================")

qpe = QuantumCircuit(3,2)

qpe.x(2)

qpe.h(0)
qpe.h(1)

qpe.cp(np.pi/2,0,2)
qpe.cp(np.pi,1,2)

qpe.measure([0,1],[0,1])

print(qpe.draw())

try:
    qpe.draw("mpl")
    plt.show()
except:
    print("Matplotlib circuit visualization unavailable")


# ==========================================================
# 9. STATEVECTOR VISUALIZATION
# ==========================================================

print("\n==============================")
print("9. Statevector Probabilities")
print("==============================")

qc_sv = QuantumCircuit(2)

qc_sv.h(0)
qc_sv.cx(0,1)

state = Statevector.from_instruction(qc_sv)

probs = state.probabilities_dict()

plt.figure(figsize=(6,4))
plt.bar(probs.keys(), probs.values())

plt.xlabel("Quantum State")
plt.ylabel("Probability")
plt.title("Bell State Probability Distribution")

plt.show()


# ==========================================================
# SUMMARY
# ==========================================================

print("\nCompleted:")
print("✓ Superposition")
print("✓ Bloch Sphere")
print("✓ Bell State Entanglement")
print("✓ Deutsch-Jozsa")
print("✓ Grover Search")
print("✓ Amplitude Amplification")
print("✓ Quantum Fourier Transform")
print("✓ Quantum Phase Estimation")
print("✓ Statevector Visualization")
```

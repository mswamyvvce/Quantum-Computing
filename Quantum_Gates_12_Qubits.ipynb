# Quantum Gate Demonstration for 12 Qubits
Uses Qiskit to demonstrate common quantum gates.


```python
from qiskit import QuantumCircuit
n = 12
qc = QuantumCircuit(n)
qc.h(0)
if n > 1:
    qc.cx(0,1)
qc.x(0)
qc.y(0)
qc.z(0)
qc.s(0)
qc.t(0)
qc.rx(0.5,0)
qc.ry(0.5,0)
qc.rz(0.5,0)
print(qc)
qc.draw('mpl')
```
